```python
# Imports and general settings
!pip install joblib
!pip install scikit-learn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (confusion_matrix, classification_report, accuracy_score,
                             precision_score, recall_score, f1_score, ConfusionMatrixDisplay)
# Models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Display options
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')
RANDOM_STATE = 42
```

```python
# Data loading
# Option A: load from sklearn (recommended for reproducibility)
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target, name='species')

# Option B: if you have a CSV (uncomment & adjust path)
# df = pd.read_csv('/path/to/Iris.csv')
# X = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
# y = df['Species'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})

print("Feature sample:")
display(X.head())
print("Target distribution:")
display(y.value_counts())

```

### Quick notes:
- `iris.feature_names` are: sepal length (cm), sepal width (cm), petal length (cm), petal width (cm).
- Target labels are integers: `0` = setosa, `1` = versicolor, `2` = virginica.

```python
# Basic EDA & visualizations
print("Dataset shape:", X.shape)
display(X.describe())

# Pairplot (good for small datasets)
sns.pairplot(pd.concat([X, y], axis=1), hue='species', corner=True, diag_kind='hist')
plt.suptitle('Pairplot of features colored by species', y=1.02)
plt.show()

# Correlation heatmap
plt.figure(figsize=(6,5))
sns.heatmap(pd.concat([X, y], axis=1).corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature correlation matrix")
plt.show()
```

# Insight from EDA
- Petal length and petal width are highly correlated — they tend to separate species well.
- One species (Setosa) is linearly separable from the other two; Versicolor and Virginica overlap.
- No missing values in the classic iris dataset.


```python
# Cell 4 — Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=RANDOM_STATE)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```

### Modeling approach
We'll create a `Pipeline` that includes `StandardScaler` (important for distance-based models and SVM) and the estimator.  
We'll evaluate models with cross-validation (stratified) and then on the held-out test set.


```python
# Cell 5 — Define models in pipelines
models = {
    "KNN": Pipeline([('scaler', StandardScaler()), ('clf', KNeighborsClassifier())]),
    "LogisticRegression": Pipeline([('scaler', StandardScaler()), ('clf', LogisticRegression(random_state=RANDOM_STATE, max_iter=200))]),
    "SVM_RBF": Pipeline([('scaler', StandardScaler()), ('clf', SVC(kernel='rbf', probability=True, random_state=RANDOM_STATE))]),
    "DecisionTree": Pipeline([('clf', DecisionTreeClassifier(random_state=RANDOM_STATE))]),
    "RandomForest": Pipeline([('clf', RandomForestClassifier(random_state=RANDOM_STATE))]),
    "GradientBoosting": Pipeline([('clf', GradientBoostingClassifier(random_state=RANDOM_STATE))])
}

# Quick cross-validation performance (accuracy)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
cv_results = {}
for name, pipeline in models.items():
    scores = cross_val_score(pipeline, X_train, y_train, cv=cv, scoring='accuracy')
    cv_results[name] = scores
    print(f"{name:15s} CV Accuracy: {scores.mean():.4f} ± {scores.std():.4f}")
```

**Interpretation:**  
These cross-validated accuracies give an initial ranking of which models perform well. Next we'll train each model on the training data and evaluate on the test set, then perform hyperparameter tuning for selected models.


```python
# Cell 6 — Train on train set and evaluate on test set (baseline)
def evaluate_model(pipeline, X_test, y_test):
    y_pred = pipeline.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='macro', zero_division=0)
    rec = recall_score(y_test, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='macro', zero_division=0)
    return {"accuracy": acc, "precision": prec, "recall": rec, "f1": f1, "y_pred": y_pred}

baseline_results = {}
for name, pipeline in models.items():
    # fit on full training set
    pipeline.fit(X_train, y_train)
    res = evaluate_model(pipeline, X_test, y_test)
    baseline_results[name] = res
    print(f"=== {name} ===")
    print(f"Accuracy: {res['accuracy']:.4f}, F1 (macro): {res['f1']:.4f}")
    print(classification_report(y_test, res['y_pred'], target_names=iris.target_names))
    print()
```

### Confusion matrix for best baseline model
(Plot whichever model had the best test F1 above — we'll display multiple for clarity.)


```python
# Plot confusion matrices for all baseline models
fig, axes = plt.subplots(2, 3, figsize=(18,10))
axes = axes.ravel()
for ax, (name, res) in zip(axes, baseline_results.items()):
    ConfusionMatrixDisplay.from_predictions(y_test, res['y_pred'], display_labels=iris.target_names, ax=ax, cmap='Blues', xticks_rotation='vertical')
    ax.set_title(f"{name} (acc={res['accuracy']:.2f})")
plt.tight_layout()
plt.show()

```

## Hyperparameter tuning (GridSearchCV)
We'll tune KNN, SVM, RandomForest and GradientBoosting with a reasonable grid. Use StratifiedKFold and scoring='accuracy' (or 'f1_macro' if you prefer).


```python
# Cell 8 — Prepare parameter grids (small but meaningful)
param_grids = {
    "KNN": {
        'clf__n_neighbors': [3,5,7],
        'clf__weights': ['uniform', 'distance'],
        'clf__p': [1,2]  # Manhattan vs Euclidean
    },
    "SVM_RBF": {
        'clf__C': [0.1, 1, 10],
        'clf__gamma': ['scale', 0.1, 1]
    },
    "RandomForest": {
        'clf__n_estimators': [50, 100],
        'clf__max_depth': [None, 3, 5],
        'clf__min_samples_split':[2,4]
    },
    "GradientBoosting": {
        'clf__n_estimators': [50, 100],
        'clf__learning_rate': [0.01, 0.1],
        'clf__max_depth': [2, 3]
    }
}

```

```python
# Cell 9 — Run GridSearchCV for selected models
tuned_models = {}
for name in ['KNN','SVM_RBF','RandomForest','GradientBoosting']:
    print(f"Running GridSearchCV for {name}...")
    pipeline = models[name]  # pipeline defined earlier
    grid = GridSearchCV(pipeline, param_grids[name], cv=cv, scoring='accuracy', n_jobs=-1)
    grid.fit(X_train, y_train)
    tuned_models[name] = grid
    print(f"Best params for {name}: {grid.best_params_}")
    print(f"Best CV score: {grid.best_score_:.4f}")
    print()
```

```python
# Cell 10 — Evaluate tuned models on test set and collect metrics
tuned_results = {}
for name, grid in tuned_models.items():
    best = grid.best_estimator_
    y_pred = best.predict(X_test)
    tuned_results[name] = {
        "accuracy": accuracy_score(y_test, y_pred),
        "precision": precision_score(y_test, y_pred, average='macro', zero_division=0),
        "recall": recall_score(y_test, y_pred, average='macro', zero_division=0),
        "f1": f1_score(y_test, y_pred, average='macro', zero_division=0),
        "y_pred": y_pred,
        "model": best
    }
    print(f"{name}: Accuracy={tuned_results[name]['accuracy']:.4f}, F1={tuned_results[name]['f1']:.4f}")

```

### Compare baseline vs tuned models
We'll create a summary table to compare baseline and tuned results (macro F1 and accuracy).


```python
# Cell 11 — Summary table
records = []
for name in models.keys():
    baseline = baseline_results[name]
    tuned = tuned_results.get(name, None)
    records.append({
        "model": name,
        "baseline_acc": baseline['accuracy'],
        "baseline_f1": baseline['f1'],
        "tuned_acc": tuned['accuracy'] if tuned else np.nan,
        "tuned_f1": tuned['f1'] if tuned else np.nan
    })

summary_df = pd.DataFrame(records).sort_values(by='tuned_f1', ascending=False)
display(summary_df)

```

**Pick the best model**: Look at `tuned_f1` (or `tuned_acc`) column. Use the model with best overall balance of metrics and interpretability for your purpose.


```python
# Cell 12 — Feature importance (for tree-based models)
# RandomForest feature importance
if 'RandomForest' in tuned_models:
    rf = tuned_models['RandomForest'].best_estimator_.named_steps['clf']
    feat_imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
    print("Random Forest feature importances:")
    display(feat_imp)
    feat_imp.plot(kind='bar')
    plt.title('Random Forest Feature Importances')
    plt.ylabel('Importance')
    plt.show()

# For GradientBoosting as well
if 'GradientBoosting' in tuned_models:
    gb = tuned_models['GradientBoosting'].best_estimator_.named_steps['clf']
    feat_imp_gb = pd.Series(gb.feature_importances_, index=X.columns).sort_values(ascending=False)
    print("Gradient Boosting feature importances:")
    display(feat_imp_gb)
    feat_imp_gb.plot(kind='bar')
    plt.title('Gradient Boosting Feature Importances')
    plt.ylabel('Importance')
    plt.show()
```

### Visualize Decision Boundaries (2D)
We will visualize decision regions on two features (e.g., petal length and petal width) to help explain model behavior. This is just for visualization — the models used four features during training.


```python
# Cell 13 — Decision boundary helper (2D)
from matplotlib.colors import ListedColormap
def plot_decision_regions_2d(X, y, classifier, features, title):
    # features: list of two column names
    X_ = X[features].values
    # Standardize for plotting (fit transform on full subset)
    scaler = StandardScaler().fit(X_)
    X_std = scaler.transform(X_)
    clf = classifier
    # Fit classifier on the 2D subset for visualization purpose
    clf.fit(X_std, y)
    x1_min, x1_max = X_std[:, 0].min() - 1, X_std[:, 0].max() + 1
    x2_min, x2_max = X_std[:, 1].min() - 1, X_std[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.02),
                           np.arange(x2_min, x2_max, 0.02))
    Z = clf.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    cmap = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    plt.figure(figsize=(8,6))
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    scatter = plt.scatter(X_std[:,0], X_std[:,1], c=y, edgecolor='k', s=40)
    plt.xlabel(features[0])
    plt.ylabel(features[1])
    plt.legend(handles=scatter.legend_elements()[0], labels=iris.target_names)
    plt.title(title)
    plt.show()

# Example: use best RandomForest (if available) on petal features
if 'RandomForest' in tuned_models:
    rf_model = tuned_models['RandomForest'].best_estimator_.named_steps['clf']
    plot_decision_regions_2d(X_train, y_train, DecisionTreeClassifier(max_depth=3, random_state=RANDOM_STATE), 
                             ['petal length (cm)', 'petal width (cm)'],
                             'Decision boundaries (toy demo with a small DecisionTree)')
else:
    print("RandomForest not tuned — skip decision region demo.")
```
